Conducted an extensive literature review where w¡one section focusses on Remote Sensing applications. Some interesting conclusions are:

- Zarate-Valdez et al. [35] focused on predicting the leaf area index (LAI) in almond orchards using VIs derived from Landsat imagery. The results show that the EVI is the most accurate index for predicting LAI, with an R2 of 0.78. 
- Sandonís-Pozo et al. [36] used satellite data to estimate geometric and structural parameters. Results showed that NDVI and GNDVI had the strongest correlations with geometric and
structural parameters.
- Caras et al. [45] used RS techniques to study the impact of weed management on almond tree growth, employing UAVs with MSP cameras for data collection. Their research evaluated practices such as ground covers, mulches, and herbicides, identifying the most effective approach as combining pre-emergent herbicides and ground cover, resulting in
higher almond yields and improved quality parameters.

```python
from matplotlib import rcParams, cycler
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import os
```

```python
# import data
root = os.getcwd()
path_to_data_folder = os.path.join(root, 'Data\\Idanha\\NDVI_NDWI_TimeSeries.csv') # cloud coverage: 20 percent

data = pd.read_csv(path_to_data_folder, parse_dates=True, index_col=5, header=0)

print(data.head())
```

```python
#clean data

data
```

```python
# Add the standard deviation as a shaded region

sns.set_theme()
plt.fill_between(data.index, data["NDVI_mean"] - data["NDVI_stdDev"], data["NDVI_mean"] + data["NDVI_stdDev"], color='darkgreen', alpha=0.1, label='Std Deviation')
plt.plot(data["NDVI_mean"], color='darkgreen')

# set labels:
plt.title('NDVI regional signal')
plt.ylabel('NDVI [-]')
plt.ylim(0,1)
plt.xlabel('Time [Years]')

plt.show()
```

```python
sns.set_theme()
plt.fill_between(data.index, data["NDWI_mean"] - data["NDWI_stdDev"], data["NDWI_mean"] + data["NDWI_stdDev"], color='darkblue', alpha=0.1, label='Std Deviation')
plt.plot(data["NDWI_mean"], color='darkblue')

# set labels:
plt.title('NDWI regional signal')
plt.ylabel('NDWI [-]')
plt.ylim(-0.5,0.5)
plt.xlabel('Time [Years]')

plt.show()
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

N = 10
data = [np.logspace(0, 1, 100) + np.random.randn(100) + ii for ii in range(N)]
data = np.array(data).T
cmap = plt.cm.coolwarm
rcParams['axes.prop_cycle'] = cycler(color=cmap(np.linspace(0, 1, N)))


from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], color=cmap(0.), lw=4),
                Line2D([0], [0], color=cmap(.5), lw=4),
                Line2D([0], [0], color=cmap(1.), lw=4)]

fig, ax = plt.subplots(figsize=(10, 5))
lines = ax.plot(data)
ax.legend(custom_lines, ['Cold', 'Medium', 'Hot']);
```
