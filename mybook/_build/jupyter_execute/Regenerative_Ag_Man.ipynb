# Regenerative Agricultural Management

A key component of Regenerative Agricultural Management (RAM) in Almond cultivation on the Iberian Peninsula is growing cover crops. To understand the (hydrological) benefit of this RAM practice we summarized below a few articles on cover crop in almond farms, including current unkowns in satellite applications, and general understanding of its effects on the production.


#### Effects of Cover Crops and Drip Fertigation Regime in a Young Almond Agroecosystem

SOURCE: {cite:t}`agronomy12112606`

Almond (Prunus dulcis) is the third most widely cultivated crop in Spain and holds significant importance in Mediterranean marginal areas due to its drought tolerance. Despite Spain being the second-largest almond producer globally, water and nitrogen availability remain critical limiting factors for yield. Efficient management of these resources—through practices such as deficit fertigation and the integration of cover crops (CC)—can improve water use efficiency (WUE) and nitrogen use efficiency (NUE), while minimizing nutrient loss and protecting water bodies from contamination.

Cover crops offer several ecosystem services, including improved soil structure, erosion control, enhanced water infiltration, increased carbon sequestration, and better pest and pollination dynamics. However, results from a recent study in a newly established almond orchard reveal that CC implementation must be approached cautiously. While CC can reduce soil nitrate concentrations and prevent leaching, they did not contribute to increases in soil organic carbon (SOC) or total nitrogen, likely due to already high initial SOC levels. In fact, SOC was observed to be lower in CC-managed areas compared to bare soil (BS) plots.

Importantly, trees grown with cover crops showed reduced vegetative growth and yield compared to those under bare soil management. This was not due to water competition—tree water status actually improved with CC due to better rainwater infiltration and deeper root exploration—but rather nutritional competition, particularly involving calcium, phosphorus, iron, manganese, and zinc. These deficiencies coincided with periods of high nutrient demand such as kernel filling and rapid root development. The third year, which had the highest fruit yield overall, also revealed the most significant negative impact: CC reduced almond yield by an average of 26%, mainly due to fewer fruits per tree, associated with smaller canopy volumes.

Furthermore, reductions in irrigation (from 75–60% to 30–60% of crop evapotranspiration) led to decreased pruning biomass, trunk size, and yield in some years. Yet, a half-input "farmer treatment" (30% ETc, 50% fertilization) achieved acceptable yields—only 19% lower than full-input conditions—indicating that resource-efficient practices can still maintain commercial viability during orchard establishment.

Overall, while cover crops can contribute positively to ecosystem health and water retention, their use in young almond orchards must be finely balanced. Nutrient competition poses a significant risk to yield and growth, underscoring the need for site-specific, adaptive management strategies to ensure both sustainability and profitability.


#### Developing cover crop systems for California almonds: Current knowledge and uncertainties 

SOURCE: {cite:t}`Wauters01012023`

Recent studies highlight the value of orchard floor diversification—whether through planted cover crops or spontaneous (resident) vegetation—in enhancing ecosystem functions in almond orchards. This diversification has been shown to improve orchard multifunctionality, bolstering non-production benefits such as soil health and biodiversity without negatively affecting crop yield. As a result, it offers a promising strategy for increasing the sustainability and resilience of Mediterranean orchard systems (Fenster et al. 2021; Morugán-Coronado et al. 2020).

Cover crops have a long-standing history in Mediterranean orchards, used for centuries to support various agronomic and ecological goals. They are especially valuable because they can be strategically managed through their three core components: roots, aboveground biomass, and flowers, each contributing uniquely to orchard performance.


Roots: Enhancing Soil and Water Functions
Cover crop roots play a crucial role in improving water infiltration by breaking up compacted soils and creating stable soil structures. In almonds, their use in orchard alleys has led to measurable increases in both water infiltration rates and volumes compared to bare soil conditions (Folorunso et al. 1992). Additionally, their root systems help combat soil compaction—reducing crusting by up to 41%—and contribute to better soil aggregation and organic matter accumulation, further enhancing soil health.

A particularly important knowledge gap remains in quantifying the water balance of cover crops under modern orchard practices, including drip and micro-irrigation systems. Collaborative research with growers is needed to ensure that the benefits of cover crops are fully understood and optimized without compromising water use efficiency.

Roots and biomass also offer potential in biological pest control. Certain species, such as brassicas, release natural nematicides (via glucosinolates) that can suppress harmful nematodes, reducing the need for chemical inputs (Zasada et al. 2010; Haramoto and Gallandt 2004).


Biomass: Vegetation and Residue Management
The aboveground vegetative growth of cover crops serves several functions. First, it provides habitat and shelter for natural enemies of insect pests, supporting biological control (Landis et al. 2000). It may also interfere with pest host-location processes, disrupting insect behavior (Andow et al. 1991). Additionally, dense biomass suppresses weeds through competition and shading, while cover crop management activities themselves can assist with integrated weed management strategies.

Thus, the quality and quantity of biomass can significantly influence pest and weed dynamics, making it a critical factor in system-level management.

Flowers: Supporting Pollinators
Flowering cover crops expand nectar and pollen resources for pollinators like honey bees during periods when almond trees are not in bloom. Research in California shows that introducing flowering plants near orchards did not impair crop pollination (Lundin et al. 2017), indicating that these species can enhance pollinator support without compromising almond productivity.

Nonetheless, to protect beneficial insects, pest management and flowering cover crops must be co-managed—especially to reduce non-target impacts from pesticide use (May et al. 2015). Combining high-biomass species (for weed suppression) with flowering species (for pollinators) offers synergistic benefits, as mixtures can fulfill multiple ecosystem service roles better than single-species cover crops.

#### Remote Sensing Applications in Almond Orchards: A Comprehensive Systematic Review of Current Insights, Research Gaps, and Future Prospects []

SOURCE: {cite:t}`app14051749`

This paper includes a literature review with a focus on remote sensing (RS) in almond orchards. Key findings include:

- Zarate-Valdez et al. used Landsat-based vegetation indices to predict leaf area index (LAI), finding that EVI was the most accurate (R² = 0.78).

- Sandonís-Pozo et al. found that NDVI and GNDVI had the strongest correlation with almond tree geometric and structural traits using satellite data.

- Caras et al. used UAVs to assess the effects of weed management strategies, concluding that a combination of pre-emergent herbicides and ground cover was the most effective, boosting yield and fruit quality.

## Example Almond Cover Crop

To get a clear idea on what almond spectrally looks like we first have to analyze what happens in Idanha and its landuse. For this exercise we consulted https://www.ifap.pt/isip/ows/ and obtained LU shapefiles from 2022 and 2024 to see where Almonds are grown and retrieve field-level satellite data, starting with Sentinel-2. We will start by looking at the regional NDVI signal of Idanha-a-Nova, a municipality within the district of Castelo-Branco. Then we look at the quality of the field-level timeseries and try to use the timeseries to get a almond growth curve and spectral signatures in and outside the growing season.

Important to mention is that most modern plantations on the Iberian Peninsula have varieties selected in Spanish, French or American breeding programs. These varieties have a wide range of characteristics with regard to flowering time and characteristics, but generally flowering takes place in March and harvesting at the beginning of August or September to avoid associated harvesting problems with precipitation.

````python
from matplotlib import rcParams, cycler
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import geopandas as gpd
from shapely.geometry import shape
import json  # or use json.loads for GeoJSON
import folium
import os
````

````python
# import data
root = os.getcwd()
path_to_data_folder = os.path.join(root, 'Data\\Idanha\\NDVI_NDWI_TimeSeries.csv') # cloud coverage: 20 percent

data = pd.read_csv(path_to_data_folder, parse_dates=True, index_col=5, header=0)
````

````python
# clean data
data = data[data["NDVI_mean"]>0.1] #Quick and dirty elimination of some cloudy images
````

````python
# Add the standard deviation as a shaded region
sns.set_theme()
plt.fill_between(data.index, data["NDVI_mean"] - data["NDVI_stdDev"], data["NDVI_mean"] + data["NDVI_stdDev"], color='darkgreen', alpha=0.1, label='Std Deviation')
plt.plot(data["NDVI_mean"], color='darkgreen')

# set labels:
plt.title('NDVI regional signal')
plt.ylabel('NDVI [-]')
plt.ylim(0,1)
plt.xlabel('Time [Years]')

plt.show()
````

````python
sns.set_theme()
plt.fill_between(data.index, data["NDWI_mean"] - data["NDWI_stdDev"], data["NDWI_mean"] + data["NDWI_stdDev"], color='darkblue', alpha=0.1, label='Std Deviation')
plt.plot(data["NDWI_mean"], color='darkblue')

# set labels:
plt.title('NDWI regional signal')
plt.ylabel('NDWI [-]')
plt.ylim(-0.5,0.5)
plt.xlabel('Time [Years]')

plt.show()
````

````python
# import data
path_to_data_folder = os.path.join(root, 'Data\\Idanha\\NDVI_TimeSeries_PerFeature_almonds_2022.csv') # cloud coverage: 20 percent
path_to_data_folder_NDWI = os.path.join(root, 'Data\\Idanha\\NDWI_TimeSeries_PerFeature_almonds_2022.csv')

data_fields_NDVI = pd.read_csv(path_to_data_folder, index_col=3, parse_dates=True,  header=0)
data_fields_NDWI = pd.read_csv(path_to_data_folder_NDWI, index_col=3, parse_dates=True,  header=0)

# clean data
data_fields_NDVI = data_fields_NDVI[data_fields_NDVI["NDVI_mean"]>0.1]
````

### Regional signal NDVI

For the NDVI signal we see after 2022 a significant drop in average signal, while the NDWI minima after 2022 seems to increase. The NDVI signal gives a clear seasonality, peaking around the end and beginning of the next year. 2022-23 were known to be drought years and it can be that we see this in the seasonality and hence the drop in NDVI. Interestingly, the minima of the NDWI go up in these drought years which may indicate wetter surfaces. We have identified a construction of a large dam in the area and hypothesize the increase in minima is due to the extra surface water and associated increase in irrigation. Something to keep in mind for further analyses.


````python

fig, ax = plt.subplots(figsize=(10, 5))
data_fields_NDVI.groupby('feature_id')['NDVI_mean'].plot(legend='True', ax=ax)
# set labels:
plt.title('NDVI field-level signal')
plt.ylabel('NDVI [-]')
ax.get_legend().remove()
# plt.legend(bbox_to_anchor=(1.04, 1), borderaxespad=0)

plt.ylim(0,1)
plt.xlabel('Time [Years]')
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(10, 5))
data_fields_NDWI.groupby('feature_id')['NDWI_mean'].plot(legend='True', ax=ax)
# set labels:
plt.title('NDWI field-level signal')
plt.ylabel('NDWI [-]')
ax.get_legend().remove()
# plt.legend(bbox_to_anchor=(1.04, 1), borderaxespad=0)

plt.ylim(-0.4,0.6)
plt.xlabel('Time [Years]')
plt.show()
````

### Almond signal NDVI

For the Almond signal in 2022 we see a rather messy optical signal on field level, indicating that the National Landuse maps may be of poor quality or that the varieties have a very distinct growing pattern creating a challenge in obtaining a clear Almond growth curve. Interestingly, we see a similar pattern as in the regional signal: after 2022 a drop in NDVI and NDWI minima going up. To continue, we focus on the VeraCruz sites, because we are 100 percent certain of the presence of Almond orchards here (https://veracruzalmonds.com/en/).

````python
# Visualize only VeraCruz
# import data
path_to_data_folder_vc = os.path.join(root, 'Data\\Idanha\\NDVI_TimeSeries_PerFeature_vc.csv') # cloud coverage: 20 percent
path_to_data_folder_NDWI_vc = os.path.join(root, 'Data\\Idanha\\NDWI_TimeSeries_PerFeature_vc.csv')

data_fields_NDVI_vc = pd.read_csv(path_to_data_folder_vc, index_col=3, parse_dates=True,  header=0)
data_fields_NDWI_vc = pd.read_csv(path_to_data_folder_NDWI_vc, index_col=3, parse_dates=True,  header=0)

# clean data
data_fields_NDVI_vc = data_fields_NDVI_vc[data_fields_NDVI_vc["NDVI_mean"]>0.1]
````

````python
# Visualize location
print(data_fields_NDVI_vc.head())

# STEP 2: Parse geometries — assuming '.geo' column contains WKT strings
data_fields_NDVI_vc['geometry'] = data_fields_NDVI_vc['.geo'].apply(lambda x: shape(json.loads(x)))

# STEP 3: Convert to GeoDataFrame
gdf = gpd.GeoDataFrame(data_fields_NDVI_vc, geometry='geometry')

# STEP 4: Drop duplicates to get unique fields by geometry
# If 'feature_id' column exists, include it for labeling
unique_fields = gdf.drop_duplicates(subset='geometry')[['feature_id', 'geometry']].reset_index(drop=True)

# STEP 5: Get centroid for each geometry for label placement
unique_fields['centroid'] = unique_fields.geometry.centroid

# STEP 6: Create a Folium map centered on average location
map_center = [unique_fields.centroid.y.mean(), unique_fields.centroid.x.mean()]
m = folium.Map(location=map_center, zoom_start=12)

# STEP 7: Add each unique field and label to the map
for _, row in unique_fields.iterrows():
    geo_json = folium.GeoJson(row.geometry.__geo_interface__)
    geo_json.add_to(m)

    # Add label at the centroid
    folium.Marker(
        location=[row.centroid.y, row.centroid.x],
        icon=folium.DivIcon(html=f"<div style='font-size:10pt'>{row['feature_id']}</div>")
    ).add_to(m)

# STEP 8: Display map
m
````

````python
sel = [2243627, 2243438, 2243368, 2243384, 2243364]
df_sel = data_fields_NDVI_vc.loc[data_fields_NDVI_vc['feature_id'].isin(sel)]
df_sel_ndwi = data_fields_NDWI_vc.loc[data_fields_NDWI_vc['feature_id'].isin(sel)]
print(df_sel.head())
````

````python
fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(10, 10), sharex=True)

# NDVI plot
df_sel.groupby('feature_id')['NDVI_mean'].plot(legend=True, ax=ax1)
ax1.set_title('NDVI Vera Cruz - Large fields southeast')
ax1.set_ylabel('NDVI [-]')
ax1.get_legend().remove()
ax1.set_ylim(0, 1)

# NDWI plot
df_sel_ndwi.groupby('feature_id')['NDWI_mean'].plot(legend=True, ax=ax2)
ax2.set_title('NDWI Vera Cruz - Large fields southeast')
ax2.set_ylabel('NDWI [-]')
ax2.set_xlabel('Time [Years]')
ax2.set_ylim(-0.4, 0.6)  # Adjust as needed based on your data

plt.tight_layout()
plt.show()
````

### Vera Cruz Almond signal NDVI

Focusing on the Almond signal of Veracruz we do see a clear almond growth pattern with NDVI. In the drought years we see a decrease in average NDVI, but also an increase in minima in the winter period in both the NDWI and NDVI. Is this related to the increase in irrigation or did the orchards adopt the planting of cover crops during winter time?
