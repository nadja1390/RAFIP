#### Remote Sensing Applications in Almond Orchards: A Comprehensive Systematic Review of Current Insights, Research Gaps, and Future Prospects [@app14051749]

Conducted an extensive literature review where w¡one section focusses on Remote Sensing applications. Some interesting conclusions are:

- Zarate-Valdez et al. [35] focused on predicting the leaf area index (LAI) in almond orchards using VIs derived from Landsat imagery. The results show that the EVI is the most accurate index for predicting LAI, with an R2 of 0.78. 
- Sandonís-Pozo et al. [36] used satellite data to estimate geometric and structural parameters. Results showed that NDVI and GNDVI had the strongest correlations with geometric and
structural parameters.
- Caras et al. [45] used RS techniques to study the impact of weed management on almond tree growth, employing UAVs with MSP cameras for data collection. Their research evaluated practices such as ground covers, mulches, and herbicides, identifying the most effective approach as combining pre-emergent herbicides and ground cover, resulting in
higher almond yields and improved quality parameters.

Trees with cover crop also reduced growth and yield compared to bare soil management. This was due to a nutritional competition, mainly of Ca, Fe, Mn and Zn, rather than to a decline in tree water status. Cover crops sequester up to 1 t/ha/year of carbon but do not increase

Deﬁcit fertigation and the use of cover crops can be effective practices to preserve and save water and nutrient resources in Mediterranean agroecosystems, but should be established with caution so as not to compromise the

Spanish almond production is the second largest in the world. Almond (Prunus dulcis Mill. (D.A. Webb)) is the third most extended crop in Spain and, due to its ability to tolerate drought, it is of enormous importance for marginal areas throughout the Mediterranean
Sea basin. Resource availability, speciﬁcally water and nitrogen, is the main factor limiting crop yield [1]. 

Managing water and nutrient resources (fertigation) together can considerably improve WUE and NUE, while reducing the potential loss of nutrients from the agroecosystem, which could pollute surface and ground water bodies [16].

Cover crops provide ecosystem services that favor erosion control, soil fertility, water inﬁltration, carbon stocks, pollination or
pest control [17,21–26]. Nowadays, combining proper resource management (water and nutrients) with proper CC management is absolutely necessary for the sustainability of the agroecosystem, since ultimately CC could compete for water and nutrients with the main crop [17,27].

Total soil carbon was not affected by fertigation dose in any soil depth, but it was affected by soil management, being lower in areas with CC

The results obtained in this newly established almond orchard indicate that it is possible to obtain competitive yields with signiﬁcant water and nutrient savings

A reduction of irrigation from moderate (75–60% ETc) to severe (60–30% of ETc) reduced pruning dry weight (in two out of six year),  runk-cross-sectional area, kernel yield and crop load [45]. In another study, in agreement with our results, a decrease of 30% in
water supply decreased the number of fruits per tree but not the individual kernel dry weight [49].

In this regard, our results consistently show a decrease in the almond vegetative growth (trunk perimeter and canopy cover) of the trees with cover crops (CC) inter-rows compared with trees with bare soil (BS) !!!

In addition, in the third year, when there was high fruit yield, the CC decreased fruit yield by 26% on average compared to the BS treatments. This decrease in yield due to CC was caused by a decrease in the number of fruits per tree—related to canopy volume—which translated into a reduction in WUE

Morugan-Coronado et al. [59] concluded that under warm and arid conditions yield could be detrimentally affected under no-tillage conditions. Our results corroborate those conclusions and point out that the CC effects on the young almond trees’ growth and yield were due to nutritional rather than water competition

This indicates that, contrary to what was expected, at the third year, there was no water competition, but an improvement of tree water status when using CC. This could be explained by improvements in rainwater inﬁltration and also by changes in root absorption zones, i.e., root exploration of deeper soil layers due to competition with CC [23,33,59].

Regarding macronutrients, there was a small but signiﬁcant decrease at the end of June in P and a larger decrease in Ca in the leaves of the trees with CC compared to the BS trees (Figure 2). Regarding microelements (Fe, Mn and Zn), there was also a decrease in the CC
trees, especially at the end of June (Figure 3). The decrease in macro and micronutrients in trees in response to CC matched periods with higher nutrient demand and nutrient uptake, the kernel ﬁll and rapid root growth [50,51].

Cover crops (grasses and legumes) after three years neither increased SOC nor the nitrogen content in the soil, which contradicts other studies under Mediterranean and semiarid conditions, where CC increased SOC [25,29,37,67]. --> can be attributed to a already high initial level of SOC.

During the almond tree establishment, the half farmer treatment, irrigated at 30% ETc and fertilized at 50% of crop requirements, obtained satisfactory yields (−19%) compared to the farmer treatment, irrigated at 60% ETc and fully fertilized.

#### Developing cover crop systems for California almonds: Current knowledge and uncertainties {cite:t}`Wauters01012023`

Recent studies in almond demonstrate that diversification, including planted or spontaneous (resident) vegetation in orchard alleys, can improve multifunctionality by enhancing nonproduction functions in the orchard without reducing crop yield, thereby providing opportunities to enhance sustainability and resilience (CitationFenster et al. 2021; CitationMorugán-Coronado et al. 2020).

Cover crops have been used in some Mediterranean orchards to accomplish a variety of goals for centuries

Cover crops have three main components that can be managed for potential benefit in the orchard system: roots, aboveground vegetative biomass,
and flowers.

ROOTS:
- Enhancing Water Infiltration. Cover crop roots improve water infiltration by physically loosening soil and building stable soil structure pores and channels (Araya et al. 2022). In almonds, cover cropping in orchard alleys can lead to increased water infiltration amounts and rates compared to bare soils (Folorunso et al. 1992).
- IMPORTANT: Quantifying cover crop water balance and the underlying balance under modern orchard production practices, especially high-effi-
ciency irrigation systems, remains a critical knowledge gap that should be addressed in collaboration with growers.
- Cover crops can contribute to improved soil structure and increased soil organic matter.
- In almond orchards, soil compaction and crusting decreased by 41% with cover crops (Folorunso et al. 1992), and in prune (Prunus domestica), cover crops increased resistance of macroaggregates to slaking (Roberson et al. 1991).
- Cover crop roots and aboveground biomass can both suppress pest nematodes by releasing natural nematicides (Zasada et al. 2010), especially the by-products of glucosinolates from brassicas (Haramoto and Gallandt 2004).

BIOMASS: VEGETATION AND RESIDUE QUALITY
- Cover crop aboveground biomass can help with management of aboveground insect pests by providing habitat for natural enemies (Landis et al. 2000) and/or by diluting host-location cues (Andow et al. 1991).
- Cover crops can suppress weeds through direct competition or by reducing weed seed germination, and can facilitate weed management because
the activities of cover crop management can also manage weeds.

FLOWERS: NECTAR AND POLLEN RESOURCES
- Flowering cover crops can expand access to nutritional resources for honey bees. Research on orchard-adjacent wildflower plantings in California showed no negative effects on crop pollination (Lundin et al. 2017), suggesting that cover crops can benefit bees without negatively affecting almond pollination.
- Therefore, cover crops and chemical pest management must be comanaged to limit pesticide exposure and injury risk for nontarget insects (May
et al. 2015).

flowering species support pollinators, and high biomass will outcompete weeds, so a combinationthat has both features can provide more benefit than a single species
Other constraints may be outside the biophysical realm, such as documented among citrus growers, where resistance to cover crops has been shown to stem from norms about the benefits of a “clean” or “tidy” orchard and assumptions that cover crops confer societal, but not farm-
level, benefits (Cerdà et al. 2018).

## Example Almond Cover Crop

````python
from matplotlib import rcParams, cycler
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import geopandas as gpd
from shapely.geometry import shape
import json  # or use json.loads for GeoJSON
import folium
import os
````

````python
# import data
root = os.getcwd()
path_to_data_folder = os.path.join(root, 'Data\\Idanha\\NDVI_NDWI_TimeSeries.csv') # cloud coverage: 20 percent

data = pd.read_csv(path_to_data_folder, parse_dates=True, index_col=5, header=0)

````

````python
# clean data
data = data[data["NDVI_mean"]>0.1] #Quick and dirty elimination of cloudy images

print(data.head())
````

````python
# Add the standard deviation as a shaded region
sns.set_theme()
plt.fill_between(data.index, data["NDVI_mean"] - data["NDVI_stdDev"], data["NDVI_mean"] + data["NDVI_stdDev"], color='darkgreen', alpha=0.1, label='Std Deviation')
plt.plot(data["NDVI_mean"], color='darkgreen')

# set labels:
plt.title('NDVI regional signal')
plt.ylabel('NDVI [-]')
plt.ylim(0,1)
plt.xlabel('Time [Years]')

plt.show()
````

````python
sns.set_theme()
plt.fill_between(data.index, data["NDWI_mean"] - data["NDWI_stdDev"], data["NDWI_mean"] + data["NDWI_stdDev"], color='darkblue', alpha=0.1, label='Std Deviation')
plt.plot(data["NDWI_mean"], color='darkblue')

# set labels:
plt.title('NDWI regional signal')
plt.ylabel('NDWI [-]')
plt.ylim(-0.5,0.5)
plt.xlabel('Time [Years]')

plt.show()
````

````python
# import data
path_to_data_folder = os.path.join(root, 'Data\\Idanha\\NDVI_TimeSeries_PerFeature.csv') # cloud coverage: 20 percent
path_to_data_folder_NDWI = os.path.join(root, 'Data\\Idanha\\NDWI_TimeSeries_PerFeature.csv')

data_fields_NDVI = pd.read_csv(path_to_data_folder, index_col=3, parse_dates=True,  header=0)
data_fields_NDWI = pd.read_csv(path_to_data_folder_NDWI, index_col=3, parse_dates=True,  header=0)

# clean data
data_fields_NDVI = data_fields_NDVI[data_fields_NDVI["NDVI_mean"]>0.1]
````

````python

fig, ax = plt.subplots(figsize=(10, 5))
data_fields_NDVI.groupby('feature_id')['NDVI_mean'].plot(legend='True', ax=ax)
# set labels:
plt.title('NDVI field-level signal')
plt.ylabel('NDVI [-]')
plt.legend(bbox_to_anchor=(1.04, 1), borderaxespad=0)

plt.ylim(0,1)
plt.xlabel('Time [Years]')
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(10, 5))
data_fields_NDWI.groupby('feature_id')['NDWI_mean'].plot(legend='True', ax=ax)
# set labels:
plt.title('NDWI field-level signal')
plt.ylabel('NDWI [-]')
plt.legend(bbox_to_anchor=(1.04, 1), borderaxespad=0)

plt.ylim(-0.4,0.6)
plt.xlabel('Time [Years]')
plt.show()
````

````python
# Visualize only VeraCruz
# import data
path_to_data_folder_vc = os.path.join(root, 'Data\\Idanha\\NDVI_TimeSeries_PerFeature_vc.csv') # cloud coverage: 20 percent
path_to_data_folder_NDWI_vc = os.path.join(root, 'Data\\Idanha\\NDWI_TimeSeries_PerFeature_vc.csv')

data_fields_NDVI_vc = pd.read_csv(path_to_data_folder_vc, index_col=3, parse_dates=True,  header=0)
data_fields_NDWI_vc = pd.read_csv(path_to_data_folder_NDWI_vc, index_col=3, parse_dates=True,  header=0)

# clean data
data_fields_NDVI_vc = data_fields_NDVI_vc[data_fields_NDVI_vc["NDVI_mean"]>0.1]
````

````python
# Visualize location
print(data_fields_NDVI_vc.head())

# STEP 2: Parse geometries — assuming '.geo' column contains WKT strings
data_fields_NDVI_vc['geometry'] = data_fields_NDVI_vc['.geo'].apply(lambda x: shape(json.loads(x)))

# STEP 3: Convert to GeoDataFrame
gdf = gpd.GeoDataFrame(data_fields_NDVI_vc, geometry='geometry')

# STEP 4: Drop duplicates to get unique fields by geometry
# If 'feature_id' column exists, include it for labeling
unique_fields = gdf.drop_duplicates(subset='geometry')[['feature_id', 'geometry']].reset_index(drop=True)

# STEP 5: Get centroid for each geometry for label placement
unique_fields['centroid'] = unique_fields.geometry.centroid

# STEP 6: Create a Folium map centered on average location
map_center = [unique_fields.centroid.y.mean(), unique_fields.centroid.x.mean()]
m = folium.Map(location=map_center, zoom_start=12)

# STEP 7: Add each unique field and label to the map
for _, row in unique_fields.iterrows():
    geo_json = folium.GeoJson(row.geometry.__geo_interface__)
    geo_json.add_to(m)

    # Add label at the centroid
    folium.Marker(
        location=[row.centroid.y, row.centroid.x],
        icon=folium.DivIcon(html=f"<div style='font-size:10pt'>{row['feature_id']}</div>")
    ).add_to(m)

# STEP 8: Display map
m
````

````python
sel = [2243627, 2243438, 2243368, 2243384, 2243364]
df_sel = data_fields_NDVI_vc.loc[data_fields_NDVI_vc['feature_id'].isin(sel)]
df_sel_ndwi = data_fields_NDWI_vc.loc[data_fields_NDWI_vc['feature_id'].isin(sel)]
print(df_sel.head())
````

````python
fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(10, 10), sharex=True)

# NDVI plot
df_sel.groupby('feature_id')['NDVI_mean'].plot(legend=True, ax=ax1)
ax1.set_title('NDVI Vera Cruz - Large fields southeast')
ax1.set_ylabel('NDVI [-]')
ax1.get_legend().remove()
ax1.set_ylim(0, 1)

# NDWI plot
df_sel_ndwi.groupby('feature_id')['NDWI_mean'].plot(legend=True, ax=ax2)
ax2.set_title('NDWI Vera Cruz - Large fields southeast')
ax2.set_ylabel('NDWI [-]')
ax2.set_xlabel('Time [Years]')
ax2.set_ylim(-0.4, 0.6)  # Adjust as needed based on your data

plt.tight_layout()
plt.show()
````
